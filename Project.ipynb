This is a deep learning based numerical character reader , this model will detect the hand writen or poorly writen number with the help of alrady available MNIST data.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import  numpy as np
import pandas as pd

```

```python
(x_train , y_train),(x_test, y_test)= mnist.load_data()



x_train= x_train.astype("float32")/255
x_test= x_test.astype("float32")/255


x_train = x_train.reshape(-1, 28, 28, 1)
x_test  = x_test.reshape(-1, 28, 28, 1)



```

Data augumentation

1.Rotation
2.Flipping
3.Zoom
4.Brightness
5.Shift


```python
datagen= ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

datagen.fit(x_train)
```

```python
from tensorflow.keras import models, layers

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),
    
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(
    datagen.flow(x_train, y_train, batch_size=32),
    epochs=10
)
```

```python
model.evaluate(x_test, y_test)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "./Untitled.png"

def predict_digit(img_path):

    img = image.load_img(img_path, color_mode="grayscale", target_size=(28,28))
    
    img_array = image.img_to_array(img)
    
    img_array = img_array / 255.0

    img_array = 1 - img_array   # invert colors

    img_array = img_array.reshape(1,28,28,1)

    prediction = model.predict(img_array)

    predicted_digit = np.argmax(prediction)

    print("Predicted Digit:", predicted_digit)


predict_digit(img_path)
```
